```python
pdf = FPDF()
pdf.add_page()
pdf.add_font('DejaVu', '', 'dejavu-fonts-ttf-2.37/ttf/DejaVuSansCondensed.ttf', uni=True)
pdf.set_font('DejaVu', '', 8)
pdf.ln(10)

for line in stripped[:3]:
    tt = ",".join(line)
    pdf.multi_cell(h=5.0, align='L', w=0, txt=tt, border=0)

pdf.output('r5.pdf', 'F')
```
